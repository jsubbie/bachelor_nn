```python
import os 
import csv
import pandas as pd 
import numpy as np
```

```python
bachelor = pd.read_csv("bachelor-contestants.csv", encoding = "ISO-8859-1")
bachelor.head()
```

```python
X = bachelor.drop("ElimWeek", axis=1)
y = bachelor["ElimWeek"]
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
```

```python
#Label-encode Occupation column
label_encoder = LabelEncoder()
label_encoder.fit(X["Occupation"])
encoded_X_occ = label_encoder.transform(X["Occupation"])

#Label-encode Hometown column
label_encoder = LabelEncoder()
label_encoder.fit(X["Hometown"])
encoded_X_hom = label_encoder.transform(X["Hometown"])

#
label_encoder = LabelEncoder()
label_encoder.fit(X["State or Country"])
encoded_X_sta = label_encoder.transform(X["State or Country"])
```

```python
for label, original_class in zip(encoded_X_occ, X["Occupation"]):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
for label, original_class in zip(encoded_X_hom, X["Hometown"]):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```
