```python
import os 
import csv
import pandas as pd 
import numpy as np
```

```python
bachelor = pd.read_csv("bachelor-contestants.csv", encoding = "ISO-8859-1")
bachelor.head()
```

```python
bachelor_df = bachelor.drop(columns=['Name'])
```

```python
X = bachelor_df.drop("ElimWeek", axis=1)
y = bachelor_df["ElimWeek"]
print(X.shape, y.shape)
```

```python
# import SKlearn model and preprocessing kits 

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from keras.utils import to_categorical
```

```python
#Label-encode Occupation column
label_encoder = LabelEncoder()
label_encoder.fit(X["Occupation"])
X2 = X
encoded_X_occ = label_encoder.transform(X["Occupation"])
X2["Occupation"] = encoded_X_occ

#Label-encode Hometown column
label_encoder = LabelEncoder()
label_encoder.fit(X["Hometown"])
X2 = X
encoded_X_hom = label_encoder.transform(X["Hometown"])
X2["Hometown"] = encoded_X_hom

#label-encode State Column
label_encoder = LabelEncoder()
label_encoder.fit(X["State"])
X2 = X
encoded_X_sta = label_encoder.transform(X["State"])
X2["State"] = encoded_X_sta
```

```python
for label, original_class in zip(encoded_X_occ, X["Occupation"]):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
for label, original_class in zip(encoded_X_hom, X["Hometown"]):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
for label, original_class in zip(encoded_X_sta, X["State"]):
    print('Original Class: ' + str(original_class))
    print('Encoded Label: ' + str(label))
    print('-' * 12)
```

```python
#One-hot encoding
one_hot_occ = to_categorical(encoded_X_occ)
one_hot_occ
```

```python
one_hot_hom = to_categorical(encoded_X_hom)
one_hot_hom
```

```python
one_hot_sta = to_categorical(encoded_X_sta)
one_hot_sta
```

```python
X_train, X_test, y_train, y_test = train_test_split(X2, y, random_state=1, stratify=y)


X_scaler = StandardScaler().fit(X_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
from keras.models import Sequential
from keras.utils import to_categorical
from keras.layers import Dense

# Create an empty sequential model
model = Sequential()

number_inputs = 3
number_hidden_nodes = 4

# First hidden layer

model.add(Dense(units=number_hidden_nodes, activation='relu', input_dim=number_inputs))

# Output layer 

number_classes = 2
model.add(Dense(units=number_classes, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model using categorical_crossentropy for the loss function, the adam optimizer,
# and add accuracy to the training metrics
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
# Use the training data to fit (train) the model
model.fit(
    X_train_scaled,
    y_train,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
# Save the model
model.save("bachelor_trained.h5")
```

```python
# Load the model
from keras.models import load_model
model = load_model("bachelor_trained.h5")
```

```python
# Evaluate the model using the training data 
model_loss, model_accuracy = model.evaluate(X_test, y_test, verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Grab just one data point to test with
test = np.expand_dims(X_test[0], axis=0)
test.shape
```

```python
# Make a prediction. The result should be 5 - STANDING
print(f"Predicted class: {model.predict_classes(test)}")
```
